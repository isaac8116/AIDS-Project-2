### Use Transcript & Video to Generate Pros & Cons
Like a regular conversation, append the transcript into the history, then ask the
model again what it thinks the pros & cons raised by the video are. The reason we
give it the transcript is to help guide the model towards using the video content
in its answer, rather than hallucinating or using its own knowledge to answer.

```python
PROMPT = f"Referring to the video and its transcript above, comprehensively analyze the pros and cons of {PRODUCT} compared to other brands."

contents.extend(
    [
        types.Content(
            role="model",
            parts=[types.Part.from_text(text=transcript)],
        ),
        types.Content(
            role="user",
            parts=[types.Part.from_text(text=PROMPT)],
        ),
    ]
)
gen_config = types.GenerateContentConfig(response_mime_type="text/plain")

analysis = ""
for chunk in client.models.generate_content_stream(
    model=MODEL,
    contents=contents,
    config=gen_config,
):
    analysis += chunk.text
    print(chunk.text, end="")
```
